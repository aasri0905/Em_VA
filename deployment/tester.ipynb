```python
import pandas as pd
import streamlit as st
import matplotlib.pyplot as plt



st.set_page_config(layout='wide',page_title="Streamlit WebRTC", page_icon="ðŸ¤–")

data = pd.read_csv("Saver.csv")
```

```python
# data_2 = data
data_2 = pd.DataFrame()
data_2['Valence'] = data['Valence']
data_2['Arousal'] = data['Arousal']
data_2['Time'] = data['X-axis']

plt.fill_between(data_2['Time'], data_2['Valence'], data_2['Arousal'], alpha=1)



# Customize the chart (optional)
plt.title('Signal Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the chart
plt.show()

# st.area_chart(data_2)
```
